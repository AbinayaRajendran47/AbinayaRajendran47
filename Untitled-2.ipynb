```python
mycursor.execute(""" SELECT e1.first_name as employee1, e2.first_name as employee2, d.department_name 
                 FROM employees e1 INNER JOIN employees e2 ON e1.department_id = e2.department_id 
                 INNER JOIN departments d ON e1.department_id = d.department_id WHERE e1.emp_id < e2.emp_id; 
""")

out = mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out, 
              headers=[i[0] for i in mycursor.description],
              showindex="always",
              tablefmt='psql'))
```
